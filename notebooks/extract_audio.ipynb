```python
def extract_wav_from_files(files):
    for file_path in files:
        import moviepy.editor as mp
        if 'csv' in file_path:
            continue
        file_name = get_file_name(file_path)

        if not file_name:
            print('extract', file_name)
            video = mp.VideoFileClip(file_path)
            new_file_path = f'data/audio/{file_name}.wav'
            video.audio.write_audiofile(new_file_path)
        else:
            print('skip', file_name)


def extract_tensors_from_files(files):
    """ extracts tensors from wav files which have csv

    Parameters
    ----------
    files : list
        list with paths to files
    """
    for file_path in files:
        if 'csv' in file_path:
            file_name = get_file_name(file_path)
            
            if not file_name:
                print('extract', file_name)
                signal, sr = load_audio(file_name, return_sr=True)
                new_file_path = f'data/audio_tensors/{file_name}.pt'
                torch.save([signal, sr], new_file_path)
            else:
                print('skip', file_name)


def extract_csv_from_files(files):
    for file_path in files:
        if 'csv' in file_path:
            file_name = get_file_name(file_path)
            if not file_name:
                print('extract', file_name)
                new_file_path = f'data/csv/{file_name}.csv'
                shutil.copy2(file_path, new_file_path)
            else:
                print('skip', file_name)
                # print('-', file_name)


def print_csv_files(files):
    for file_path in files:
        if 'csv' in file_path:
            file_name = get_file_name(file_path)
            print('-', file_name)
```
